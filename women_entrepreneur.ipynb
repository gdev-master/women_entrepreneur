```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading dataset into pandas dataframe
df = pd.read_csv("/Users/gdev15/Desktop/Women Disadvantages/Dataset3.csv", delimiter=";")
```

# Exploratory Data Analysis

```python
print(df["European Union Membership"].value_counts())
print("\n")
print(df["Level of development"].value_counts())
print("\n")
print(df["Currency"].value_counts())
```

```python
df.groupby("European Union Membership")[["Women Entrepreneurship Index", "Entrepreneurship Index", 
                                         "Inflation rate", "Female Labor Force Participation Rate"]].median()
```

As we can see, Women members of the European Union Membership have better Entrepreneur Opportunities

```python
df.groupby("Level of development")[["Women Entrepreneurship Index", "Entrepreneurship Index", 
                                         "Inflation rate", "Female Labor Force Participation Rate"]].median()
```

When the Level of Development is already developed we can see that every index is better than those that are developing.

```python
country = df.groupby("Country")[["Women Entrepreneurship Index",
                       "Entrepreneurship Index", "Inflation rate",
                       "Female Labor Force Participation Rate"]].median().sort_values(by=["Female Labor Force Participation Rate"],
                                                                                     ascending=False)
df_country = pd.DataFrame(country)
df_country
```

It is great to know that we have countries where the entrepreneurship index of the Women is more than the normal index.

```python
df_country_top10 = df_country[df_country["Women Entrepreneurship Index"] > 63]
df_country_top10
```

```python
fig, ax3 = plt.subplots(figsize=(15,10))
sns.barplot(x=df_country.index, y="Women Entrepreneurship Index", data=df_country,
            order=df_country.sort_values("Women Entrepreneurship Index", ascending=True).index, ax=ax3)
plt.xticks(rotation=90)
plt.title("Women Entrepreneurship Index by Country")
```

```python
fig, ax4 = plt.subplots(figsize=(15,10))
sns.barplot(x=df_country.index, y="Entrepreneurship Index", data=df_country,
            order=df_country.sort_values("Women Entrepreneurship Index", ascending=True).index, ax=ax4)
plt.xticks(rotation=90)
plt.title("Entrepreneurship Index by Country")
```

There are countries that their Women Entrepreneurship Index is better than the Entrepreneurship Index.
